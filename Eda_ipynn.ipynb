```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
import seaborn as sns

df = pd.read_csv('Customer Churn.csv')
df
```

```python
df.info()
```

```python
#Replacing blank with 0 as tenure is 0 and no charges are recorded.
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df["customerID"].duplicated().sum()
```

# covert o and 1 value into yes and no for seniorcitizen

```python
def conv(value):
  if value == 1:
    return "yes"
  else:
      return "no"

df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)
```

```python
df.head(10)
```

```python
ax = sns.countplot(x = 'Churn', data = df)

ax.bar_label(ax.containers[0])
pt.show()
```

```python
pt.figure(figsize = (3,4))
gb = df.groupby("Churn").agg({'Churn':"count"})
pt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")
pt.title("Percentage of Chured customeres", fontsize = 10)
pt.show()
```

FROM THE GIVEN piechart we can conclude that 26.54% of our customers have churned out

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group data and count occurrences
data = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)

# Create stacked bar chart
data.plot(
    kind='bar',
    stacked=True,
    figsize=(5, 4)
)

# Add title and labels
plt.title("Churn by SeniorCitizen(%)")
plt.xlabel("Senior Citizen")
plt.ylabel("percentage")

# Rotate x-axis labels
plt.xticks(rotation=0)

# Show legend
plt.legend(title="Churn")

# Display chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group data and count occurrences
data = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)

# Convert counts to percentages
data_percent = data.div(data.sum(axis=1), axis=0) * 100

# Create stacked bar chart
data_percent.plot(
    kind='bar',
    stacked=True,
    figsize=(5, 4)
)

# Add title and labels
plt.title("Churn by SeniorCitizen (%)")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage")

# Rotate x-axis labels
plt.xticks(rotation=0)

# Show legend
plt.legend(title="Churn")

# Display chart
plt.show()
```

comparatively a greated percentage of of people in senior citizen category have churned

```python
import matplotlib.pyplot as plt
import seaborn as sns

columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Create a 3x3 grid of subplots
fig, axes = plt.subplots(3, 3, figsize=(18, 12))

# Flatten the axes array
axes = axes.flatten()

# Loop through columns and create countplots
for i, col in enumerate(columns):
    sns.countplot(
        x=col,
        data=df,
        hue='Churn',
        ax=axes[i]
    )
    axes[i].set_title(f'Churn by {col}')
    axes[i].tick_params(axis='x', rotation=45)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))

ax = sns.countplot(x="PaymentMethod", data=df, hue = "Churn")

ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.xticks(rotation=45)
plt.title("churned customer by payment method")

plt.show()
```
